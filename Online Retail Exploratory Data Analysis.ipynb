## Check relevance of time period to Sales

```python
#convert 'InvoiceDate' to datetime for analysis
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
```

```python
#Extract hour of sales
a = df['InvoiceDate'].dt.hour.to_frame()
a.rename(columns = {'InvoiceDate' : 'Sales_Hour'}, inplace=True)
df = df.join(a)
df
```

```python
# Extract month of Sales
b = df['InvoiceDate'].dt.day_name().to_frame()
b.rename(columns = {'InvoiceDate' : 'Sales_Day'}, inplace=True)
df = df.join(b)
df
```

```python
# Extract month of Sales
c = df['InvoiceDate'].dt.month_name().to_frame()
c.rename(columns = {'InvoiceDate' : 'Sales_Month'}, inplace=True)
df = df.join(c)
df
```

### Time/Period customers engage to buy online

#### Hour

```python
plt.figure(figsize=(10,6))

gr_sales = df.groupby(['Sales_Hour']).count()
gr_sales
gr_sales.reset_index(inplace=True)


plt.plot(gr_sales.Sales_Hour, gr_sales.Sales)
plt.xticks(gr_sales.Sales_Hour)

plt.title('Online Visitations per Hour', fontdict={'fontsize':20})
plt.xlabel('Time', fontdict={'fontsize':15})
plt.ylabel('No. of Visitation', fontdict={'fontsize':15})
plt.grid()
plt.show()
```

##### Online visitation for all customers peaked at 12pm. 
##### Marketing agency might use this information to target ads before and during this period. Despite sharp decline in online presence after 12 pm
##### More online presence close to the peak value occurred aroud 3pm and started declining onwards

```python
plt.figure(figsize=(10,6))

df.groupby(['Sales_Hour'])['Sales'].sum().sort_values().plot(kind='bar')
plt.title('Amount Purchased per Time', fontdict={'fontsize':20})
plt.ylabel('Amount (USD)', fontdict={'fontsize':15})
plt.xlabel('Time', fontdict={'fontsize':15})
plt.show()
```

##### However, the highest amount of sales of items happened between 10 and 11am. This may be due to time-delivery of items issues 

### Month

```python
plt.figure(figsize=(10,6))

df.groupby(['Sales_Month'])['Sales'].sum().sort_values().plot(kind='bar', color='#7d1f12')
plt.title('Total Amount Spent per Month', fontdict={'fontsize':20})
plt.ylabel('Amount (USD)', fontdict={'fontsize':15})
plt.xlabel('Month', fontdict={'fontsize':15})
# plt.xticks(df['Sales_Month'])
plt.show()

print('Data shows increase in sales for the last four months of the year')
```
